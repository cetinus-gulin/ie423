import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

load data

```python
dfValues = pd.read_csv('final_data.csv')
dfValues.head()
```

```python
dfValues.describe(include='all')
```

Fit a Regression Model to Predict Current Value of a Player

```python
# Assign independent(Y) and dependent(X) variables
X = dfValues.loc[:,['highest_value','age', 'goals', 'assists']]
Y = dfValues.loc[:,['current_value']]
```

```python
# Fit linear regression model - aka train model
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X, Y)
```

```python
# Display regression model
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

Explore Data

Histogram for the Current Values of Players

```python
# Set data fields
y = dfValues.loc[:,'current_value']

# Set histogram parameters
binwidth = 1000000.0
binrange=np.arange(min(y), max(y) + binwidth, binwidth)

# Show plot
plt.hist(y, bins=binrange)
plt.xlabel('Current Value')
plt.ylabel('Count')
plt.show()
```

Highest Value-to-Current Value Graph

```python
# Set data fields
x = dfValues.loc[:,'highest_value']
y = dfValues.loc[:,'current_value']

# Generate trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

# Show plot
plt.scatter(x, y)
plt.plot(x,p(x),"r--")
plt.xlim(0,np.max(dfValues.loc[:,'highest_value']))
plt.ylim(0,np.max(dfValues.loc[:,'current_value']))
plt.xlabel('Highest Value')
plt.ylabel('Current Value')
plt.show()
```
